```python
link_list = df['tmdbId'].fillna(0).tolist()
```

```python
link_list = list(map(int,link_list))
```

```python
len_list = len(link_list)
```

```python
plot_df = pd.DataFrame(columns=['tmdb_id', 'plot'])
plot_df
```

```python
for num in tqdm(link_list[8712:]):
    if num == 0:
        plot_df = plot_df.append({'tmdb_id': num, 'plot' : "not applicable"} , ignore_index=True)
    else:
        url = 'https://api.themoviedb.org/3/movie/'+ str(num) +'?api_key=4b1c7165172d96b4531d9d50f7464660&language=en-US'
        response = requests.get(url)
        data = response.json()
        plot = data.get('overview')
        plot_df = plot_df.append({'tmdb_id': num, 'plot' : plot} , ignore_index=True)
```
